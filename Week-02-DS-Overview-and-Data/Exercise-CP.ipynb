### Bonus #1

This time, format the table output nicely (*$12,345.00*) without modifying the values in the `DataFrame`. That is, `df_salary` should be identical before versus after running your code.

(*Hint: if you run into an error about `jinja2` perhaps you need to `pip install` something.*)

```python
gb = df_salary_tech_usd_usa.groupby('state')
stats = gb['num_salary'].agg( ['min', 'mean', 'max'] ) # collect the stats

def format_salary(salary):
    return f'${salary:,.2f}'

stats_formatted = stats.map(format_salary) # apply custom function to format stats

stats_formatted
```

### Bonus #2

Filter out the non-single-states (e.g., `'California, Colorado'`) in the most elegant way possible (i.e., *not* by blacklisting all the bad values).

```python
#GET SINGLE STATES, Any states that have a comma in it are a multiple states, so check for ones that do not have commas
non_single_states = df_salary_usd_usa['state'].str.contains(',', na=False) #nan values will not be considered
df_salary_usd_usa_single_states = df_salary_usd_usa[~non_single_states] #negate here so that we only get single states
# df_salary_usd_usa_single_states['state'].unique() #SANITY CHECK


#GROUPBY AND APPLY FORMATTING FUNCTION
gb = df_salary_usd_usa_single_states.groupby('state')
stats = gb['num_salary'].agg( ['min', 'mean', 'max'] ) # collect the stats

def format_salary(salary):
    return f'${salary:,.2f}'

stats_formatted = stats.map(format_salary) # apply custom function to format stats

stats_formatted
```
